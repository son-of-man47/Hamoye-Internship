# QUESTION13 : MEAN ABSOLUTE ERROR

```python
##defining our model
linear_model = LinearRegression()
```

```python
##defining our scale
scaler = MinMaxScaler()
```

```python
##scaling the data
n_df = pd.DataFrame(scaler.fit_transform(df), columns= df.columns)
n_df.head()
```

```python
##selecting our x(predictors/independent variables) and y(response/dependent variable)
x = n_df.drop("Appliances", axis = 1)
y = n_df["Appliances"]
```

```python
##splitting the x and y
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
##fitting and predicting

linear_model.fit(x_train, y_train)
y_pred = linear_model.predict(x_test)
y_pred
```

CALCULATING MATRICS NEEDED

```python
mae = mean_absolute_error(y_test, y_pred).round(2)
mae
```

# QUESTION 14: RESIDUAL SUM OF SQUARES

```python
rss = np.sum(np.square(y_test - y_pred)).round(2)
rss
```

# QUESTION 15: Root Mean Squared Error

```python
rmse = np.sqrt(mean_squared_error(y_test, y_pred)).round(3)
rmse
```

# QUESTION 16: R2_SCORE 

```python
r2 = r2_score(y_test, y_pred).round(2)
r2
```

# QUESTION 17: DETERMINING FATURE WEIGHTS FOR THE LINEAR MODEL

```python
##function for determining feature weights

def get_weights_df(model, features, col_name):
    weights = pd.Series(model.coef_, ##this brings the coefficient of the model
                       features.columns).round(3).sort_values()
    weights_df = pd.DataFrame(weights).reset_index()
    weights_df.columns = ["Features", col_name]
    return weights_df
```

```python
linear_model_weights = get_weights_df(linear_model, x_train, "Linear_Model_Weights")
```

```python
linear_model_weights
```

# QUESTION 18: TRAINING A RIDGE MODEL WITH AN ALPHA OF 0.4 AND DETERMINING THE RMSE

```python
ridge = Ridge(alpha = 0.4) ##defining our model and the alpha
ridge.fit(x_train, y_train) ##fitting 
y_pred2 = ridge.predict(x_test) ##making predictions
y_pred2
```

# QUESTION 19: TRAINING A LASSO MODEL WITH AN ALPHA OF 0.001 AND OBTAINING THE FEATURE WEIGHTS

```python
lasso = Lasso(alpha = 0.001) ##defining our model
lasso.fit(x_train, y_train) ##fitting
y_pred3 = lasso.predict(x_test) ##predicting
y_pred3
```
