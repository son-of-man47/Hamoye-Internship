# DATA WRANGLING

```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/WalePhenomenon/climate_change/master/fuel_ferc1.csv"
fuel_data = pd.read_csv(url)
# fuel_data.head()
fuel_data.describe(include = "all")
```

```python
##check for na values
fuel_data.isna().sum()
```

```python
fuel_data.isnull().sum()
```

```python
fuel_data.groupby("fuel_unit")["fuel_unit"].count()
```

```python
fuel_data[["fuel_unit"]] = fuel_data[["fuel_unit"]].fillna(value = "mcf")
```

```python
fuel_data.isnull().sum()
```

```python
fuel_data.groupby( 'fuel_type_code_pudl' ).first()
```

Merging in Pandas

```python
f_df1 =  fuel_data.iloc[0:19000].reset_index(drop = True)
f_df2 = fuel_data.iloc[19000:].reset_index(drop = True)
```

```python
###checking the lengths

assert len(fuel_data) == (len(f_df1) + len(f_df2))
```

```python
##merging

pd.merge(f_df1, f_df2, how = "inner")
```

```python
pd.merge(f_df1,f_df2, how = "outer")
```

```python
pd.merge(f_df1, f_df2, how = "left")
```

```python
##checking for duplicates
fuel_data.duplicated().any()
```

# DATA VISUALIZATION 

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
plt.figure(figsize = (7,4))
plt.xticks(rotation = 90)
fuel_unit = pd.DataFrame({"unit":  ["BBL",
                                   "GAL", "GRAMSU", "KGU",
                                   "MCF", "MBTU", "MWDTH",
                                   "MWHTH", "TON"],
                         "count": [7998, 84, 464, 110,
                                  11354, 180, 95, 
                                   100, 8958]})
sns.barplot(data = fuel_unit, x = "unit", y = "count")
plt.xlabel("Fuel Unit")
```

```python
g = sns.barplot(data = fuel_unit, x ="unit", y = "count")
g.set_yscale("log") ##sets the y scale to a log scale
g.set_ylim(1, 12000)
plt.xlabel("Fuel Unit")
```
